```python
import pandas as pd
id='1kH53QByTJPUyUILvw5N1nNeRK1K15Zje'
url= f"https://docs.google.com/uc?id={id}&export=download"
df = pd.read_csv(url)
df
```

```python
#HW5.2
#1.LOL數據庫column有11個特徵和1個標籤,先讀取LOL.xlsx儲存到變數df，再將變數df從pandas轉成147x12的numpy陣列
import pandas as pd
import numpy as np
#print(df)
##df.head()
#df['名稱'] #變數df中有幾個遊戲角色名稱
#df['類型'] #變數df中有幾個遊戲角色類型
df = np.array(df)
print("#1變數df從pandas轉成147x12的numpy陣列: \n", df, "\n")
#print("here")
#2計算變數的11個特徵之平均值儲存到變數mu的 1x11 numpy陣列與其標準差儲存到變數sigma的 1x11 numpy陣列。
mu = np.mean(df[:,1:12], 0, dtype = np.float32)
print("#2.(1)變數的11個特徵之平均值: 變數 mu = \n",mu,"\n")
sigma = np.std(df[:,1:12], 0, dtype = np.float32)
print("#2.(2)變數的11個特徵之標準差: 變數 sigma = \n",sigma,"\n")
#3將變數df的每個column做標準化 (X-mu)/sigma, 並儲存到變數df_norm。
df_norm = np.zeros((147, 11))
X = np.zeros((147, 11))
Y = np.zeros((147, 1))
a = np.zeros((147,11))
i = 0
while i < 147:
  for j in range(0,11):
    a[i,j] = ((df[i,j+1] - mu[j]) / sigma[j])
    df_norm[i,j] += a[i,j]
  i = i + 1
print("#3將變數df的每個column做標準化,df_norm = \n", df_norm, "\n")
#4變數df_norm的11個特徵之147x11陣列與標籤之147x1陣列，分別儲存到變數X與變數Y。
X = df_norm
Y = df[:,12]
print("#4.(1)變數df_norm的11個特徵之147x11陣列儲存到變數 X =\n", X, "\n" )
print("#4.(2)變數df_norm的標籤之147x1陣列儲存到變數 Y =\n", Y, "\n")
#print("Y =",Y[146])
#print("len(X) = ",len(X) )
#print("len(Y) = ",len(Y) )
#5為了把所有角色拆分成「訓練、驗證、測試」數據，請將變數X與變數Y的row切分成8:1:1，
#分別儲存到X_train, X_valid, X_test與Y_train, Y_valid, Y_test。
#5.(1)變數X的row切分成8:1:1,分別儲存到X_train, X_valid, X_test。
len_X_train = int(len(X)*0.8)
#len_X_valid = len(X)*0.1
#len_X_test = len(X)*0.1
len_X_valid = int((len(X) - len_X_train) / 2)
len_X_test = len_X_valid
print("#5.(1)將變數X的row切分成8:1:1, length of X_train, X_valid, X_test = ",len_X_train, len_X_valid, len_X_test,"\n")
X_train = X[0:len_X_train,:]
X_valid = X[len_X_train : (len_X_train + len_X_valid):,:]
X_test = X[(len_X_train + len_X_valid) : len(X),:]
print("變數X的訓練數據 X_train = \n", X_train, "\n")
print("變數X的驗證數據 X_valid = \n", X_valid, "\n")
print("變數X的測試數據 X_test = \n", X_test, "\n")
#print(len(X_train), len(X_valid), len(X_test))
#5.(2)變數Y的row切分成8:1:1,分別儲存到Y_train, Y_valid, Y_test。
len_Y_train = int((len(Y)*0.8))
#print("here", len_Y_train)
len_Y_valid = int((len(Y) - len_Y_train) / 2)
len_Y_test = len_Y_valid
print("#5.(2)將變數Y的row切分成8:1:1, length of Y_train, Y_valid, Y_test = ",len_Y_train, len_Y_valid, len_Y_test, "\n")
Y_train = Y[0:len_Y_train]
Y_valid = Y[len_Y_train : (len_Y_train + len_Y_valid)]
Y_test = Y[(len_Y_train + len_Y_valid) : len(Y)]
print("變數Y的訓練數據 Y_train = \n", Y_train, "\n")
print("變數Y的驗證數據 Y_valid = \n", Y_valid, "\n")
print("變數Y的測試數據 Y_test = \n", Y_test, "\n")
#print(len(Y_train), len(Y_valid), len(Y_test))
```
